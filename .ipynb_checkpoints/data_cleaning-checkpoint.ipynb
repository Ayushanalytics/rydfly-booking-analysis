# Script to clean and preprocess Kanpur_Bookings_10000.xlsx dataset, handling missing values, data types, and outliers

```python
# Import required libraries for data manipulation, visualization, and analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset from an Excel file
df = pd.read_excel("Kanpur_Bookings_10000.xlsx")
```

```python
# Display the dimensions of the dataset (rows, columns)
df.shape
```

```python
# Display information about the dataset, including column names, data types, and non-null counts
df.info()
```

```python
# Display the first 10 rows of the dataset
df.head(10)
```

```python
# Generate summary statistics for all columns, including categorical and numerical data

df.describe(include='all')
```

```python
# Plot a histogram of Driver_Ratings

sns.histplot(df['Driver_Ratings'].dropna(), bins=10, kde=True)
plt.title("Driver Rating Distribution")
plt.show()

# Plot a histogram of Customer_Ratings
sns.histplot(df['Customer_Rating'].dropna(), bins=10, kde=True)
plt.title("Customer Rating Distribution")
plt.show()
```

```python
# Check the number of missing values in each column to identify data quality issues
df.isnull().sum()
```

```python
# Display the count of each booking status to understand the distribution of ride outcomes
df['Booking_Status'].value_counts()
```

```python
# Fill missing values in numerical columns with 0, as they likely indicate no data (e.g., canceled rides)
num_cols = ['V_TAT', 'C_TAT', 'Driver_Ratings', 'Customer_Rating']
df[num_cols] = df[num_cols].fillna(0)


# Fill missing values in categorical/text columns with 'N/A' to maintain consistency
text_cols = df.columns.difference(num_cols)
df[text_cols] = df[text_cols].fillna("N/A")
```

```python
# Verify that all missing values have been handled by checking for nulls again
df.isnull().sum()
```

```python
# Check the data types of Date and Time columns to ensure proper formatting
print(df['Date'].dtype)
print(df['Time'].dtype)
```

```python
# Convert Time column to datetime.time format for consistency and easier manipulation
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce').dt.time
```

```python
# Verify the type of the first element in the Time column to confirm successful conversion
type(df['Time'].iloc[0])
```

```python
# Generate summary statistics for key numerical columns after cleaning to review updated distributions
df[['Booking_Value', 'Ride_Distance', 'Driver_Ratings', 'Customer_Rating']].describe()
```

```python
# Check the booking status for rides with zero distance to understand why they were not completed
df[df['Ride_Distance'] == 0]['Booking_Status'].value_counts()
```

```python
# Check the booking status for rides with zero driver ratings to confirm they align with incomplete rides
df[df['Driver_Ratings'] == 0]['Booking_Status'].value_counts()
```

```python
# Check the booking status for rides with zero customer ratings to confirm they align with incomplete rides
df[df['Customer_Rating'] == 0]['Booking_Status'].value_counts()
```

```python
# Check for duplicate rows in the dataset

duplicate_rows = df[df.duplicated()]
print(f"Duplicate rows found: {duplicate_rows.shape[0]}")
```

```python
# Plot boxplots for numerical columns to identify outliers and understand data distribution
cols = ['Booking_Value', 'Ride_Distance', 'V_TAT', 'C_TAT']
for col in cols:
    plt.figure(figsize=(8, 1.5))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot for {col}')
    plt.show()
```

```python
# Save the cleaned dataset to a CSV file for further analysis
df.to_csv('cleaned_kanpur_data.csv', index=False)
```
